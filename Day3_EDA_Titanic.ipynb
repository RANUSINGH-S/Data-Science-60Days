```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Survived')
plt.title('survival count')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Fare'],bins=30,kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Survived',hue='Sex')
plt.title('Survival by Gender')
plt.show()
```

```python
Females survived more than males; survival rate ~74% vs ~19%.
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Survived',hue='Pclass')
plt.title('Survival by Passenger Class')
plt.show()
```

```python
1st class passengers had higher survival (~63%) than 3rd class (~24%).
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Age'],bins=30,kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
Children (age < 15) had slightly higher survival than adults.
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Fare'],bins=30,kde=True) 
plt.title('Fare Distribution') 
plt.show()
```

```python
Higher fare passengers had better survival, correlating with higher-class cabins.
```

```python
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
Pclass negatively correlates with Survived; Fare positively correlates with Survived.
```
