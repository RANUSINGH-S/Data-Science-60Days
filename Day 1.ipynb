```python
#Functions
def is_prime(n):
    if n<=1:
        return False
    for i in range(2, int(n**0.5) +1):
        if n%i==0:
            return False
    return True
print(is_prime(11))
```

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in s:
        if ch in vowels:
            count += 1
    return count
print(count_vowels("Data Science"))
```

```python
#Lists and Dictionaries
nums = [10,22,3,5,7,45,12,99,4,1]
print("MAx:",max(nums))
print("Min:",min(nums))
print("Sum:",sum(nums))
print("Mean:",sum(nums)/len(nums))
```
