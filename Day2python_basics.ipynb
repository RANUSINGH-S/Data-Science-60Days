```python
#Lists
numbers = [10,20,30,40,50]
print("List:",numbers)
print("First element:",numbers[0])
print("Slice:",numbers[1:4])
```

```python
#Tuples
tup = (10,20,30)
print("Tuple:",tup)
```

```python
#Sets
s = {10,20,20,30}
print("set (unique):",s)
```

```python
#Dictionaries
student = {"name": "Ranu","age":21,"branch":"AIML"}
print(student)
print(student["name"])
```

```python
#Simple functins
def square(x):
    return x*x
print(square(5))
print(square(12))
```

```python
#if-else
x=20
if x>10:
    print("x is big")
else:
    print("x is small")
```

```python
#Loops
for i in range(1, 6):
    print("Number:",i)
```

```python
#Numpy basics
import numpy as np
arr=np.array([10,20,30])
print("Array:",arr)
print("MEan:",np.mean(arr))
print("Std Dev:",np.std(arr))
```
