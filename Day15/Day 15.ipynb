Introduction to XGBoost

While Random Forest builds multiple trees independently and averages their results, Boosting is a sequential process. Each new tree is grown specifically to correct the errors made by the previous trees.

Step 1: Install and Import XGBoost
XGBoost is a powerful, optimized version of Gradient Boosting designed for speed and performance.

```python
pip install xgboost
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("WineQT.csv")
```

```python
df.head()
```

```python
x=df.drop('quality',axis=1)
y=(df['quality']>5).astype(int)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
rf_model=RandomForestClassifier(n_estimators=100,max_depth=6,random_state=42)
```

```python
rf_model.fit(x_train_scaled,y_train)
```

```python
rf_preds=rf_model.predict(x_test_scaled)
```

```python
print(f"Random Forest Accuracy:{accuracy_score(y_test,rf_preds):.4f}")
```

Step 2: Understand the "Boosting" Logic
In Boosting, models are added one at a time. Each subsequent model attempts to correct the residual errors of the existing ensemble.

Learning Rate (eta): This is a crucial hyperparameter. It scales the contribution of each new tree. A smaller learning rate makes the model more robust to overfitting but requires more trees (n_estimators).

Step 3: Train the Baseline XGBoost Model
Use the same scaled training data (x_train_scaled, y_train) that you used for your Random Forest.

```python
# Initialize the XGBClassifier
# We'll start with standard parameters similar to your tuned RF

xgb_model =XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    use_label_encode=False,
    eval_metric='logloss'
)
```

```python
# Train the model

xgb_model.fit(x_train_scaled,y_train)
```

Step 4: Evaluate and Compare
Now, see how the "Booster" stacks up against the "Forest".

Predict: Run predictions = xgb_model.predict(x_test_scaled).

Score: Calculate the accuracy using accuracy_score(y_test, predictions).
